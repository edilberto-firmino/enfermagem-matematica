## 3. Criar Mapa Base com Folium

```python
# Coordenadas centrais: Prefeito José Walter (região de interesse)
fortaleza_coords = [-3.8165, -38.5280]

# Criar mapa base com zoom aumentado para a região
mapa = folium.Map(
    location=fortaleza_coords,
    zoom_start=13,
    tiles='OpenStreetMap'
)

print("Mapa base criado com sucesso!")
print(f"📍 Centro do mapa: Prefeito José Walter")
print(f"   Latitude: {fortaleza_coords[0]}")
print(f"   Longitude: {fortaleza_coords[1]}")
```

## 4. Adicionar Marcadores de Postos de Saúde

```python
# Adicionar marcadores de postos de saúde (cor verde)
for idx, row in postos_saude.iterrows():
    folium.CircleMarker(
        location=[row['latitude'], row['longitude']],
        radius=8,
        popup=row['nome'],
        color='green',
        fill=True,
        fillColor='green',
        fillOpacity=0.7,
        weight=2,
        tooltip=f"{row['nome']} - Posto de Saúde"
    ).add_to(mapa)

print(f"{len(postos_saude)} postos de saúde adicionados ao mapa!")
```

## 5. Adicionar Marcadores de Postos Policiais

```python
# Adicionar marcadores de postos policiais (cor azul)
for idx, row in postos_policiais.iterrows():
    folium.CircleMarker(
        location=[row['latitude'], row['longitude']],
        radius=8,
        popup=row['nome'],
        color='blue',
        fill=True,
        fillColor='blue',
        fillOpacity=0.7,
        weight=2,
        tooltip=f"{row['nome']} - Delegacia"
    ).add_to(mapa)

print(f"{len(postos_policiais)} postos policiais adicionados ao mapa!")
```

## 6. Adicionar Marcadores de CAPS

```python
# Adicionar marcadores de CAPS (cor roxo)
for idx, row in caps.iterrows():
    folium.CircleMarker(
        location=[row['latitude'], row['longitude']],
        radius=8,
        popup=row['nome'],
        color='purple',
        fill=True,
        fillColor='purple',
        fillOpacity=0.7,
        weight=2,
        tooltip=f"{row['nome']} - CAPS"
    ).add_to(mapa)

print(f"{len(caps)} CAPS adicionados ao mapa!")
```

## 7. Adicionar Marcadores de Corpo de Bombeiros

```python
# Adicionar marcadores de corpo de bombeiros (cor vermelho)
for idx, row in corpo_bombeiros.iterrows():
    folium.CircleMarker(
        location=[row['latitude'], row['longitude']],
        radius=8,
        popup=row['nome'],
        color='red',
        fill=True,
        fillColor='red',
        fillOpacity=0.7,
        weight=2,
        tooltip=f"{row['nome']} - Bombeiros"
    ).add_to(mapa)

print(f"{len(corpo_bombeiros)} postos de bombeiros adicionados ao mapa!")
```

## 8. Gerar Mapa de Calor com Heatmap

```python
# Preparar dados para o heatmap (lista de [lat, lon, intensidade])
# A intensidade é baseada na densidade de locais
dados_heatmap = []

# Adicionar todos os pontos com intensidade 1 (aumentará automaticamente por proximidade)
for idx, row in todos_dados.iterrows():
    dados_heatmap.append([row['latitude'], row['longitude'], 1])

# Criar e adicionar o heatmap
HeatMap(dados_heatmap, radius=20, blur=15, max_zoom=13).add_to(mapa)

print(f"Mapa de calor criado com {len(dados_heatmap)} pontos!")
```

## 9. Salvar e Exibir o Mapa

```python
# Salvar o mapa em arquivo HTML
mapa.save('mapa_calor_fortaleza.html')
print("Mapa salvo em: mapa_calor_fortaleza.html")

# Exibir o mapa no notebook
display(mapa)
```

## Resumo

Este mapa de calor mostra a distribuição e proximidade de locais de segurança e saúde em Fortaleza:

- **Verde**: Postos de Saúde
- **Azul**: Delegacias/Postos Policiais  
- **Roxo**: CAPS (Centros de Atenção Psicossocial)
- **Vermelho**: Corpos de Bombeiros

A camada de mapa de calor (heatmap) destaca as áreas com maior concentração de serviços. As cores mais quentes (amarelo/vermelho) indicam maior densidade de locais próximos. Você pode clicar nos marcadores para ver mais informações sobre cada local.
